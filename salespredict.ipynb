# univariate analysis

```python
for i in df123.select_dtypes(include="number"):
    sns.boxplot(df123[i])
    plt.show()
```

```python
sns.histplot(df123, bins=10, color='black', kde=False)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Plot')
plt.grid(True)
plt.show()
```

```python
for i in df123.select_dtypes(include="number"):
    skew=df123[i].skew()
    print("the skewness of",i,"is",skew)
    print()
```

# Bi variate analysis


```python
df123.columns
```

```python
fig, axes = plt.subplots (3,2, figsize=(18,12))
ax_index = [(i,j) for i in range(3) for j in range(2)]
index_number = 0
fig.suptitle('Effect of various factors on Weekly Sales', fontsize=18, color = '#06917e', y = 1.05)
for i in ['Unemployment', 'IsHoliday', 'Size', 'CPI', 'Temperature', 'Fuel_Price']:
    sns.scatterplot(x=i, y='Weekly_Sales', data=df123, ax=axes[ax_index[index_number]], palette='afmhot_r')
    index_number += 1
    plt.tight_layout()
```
