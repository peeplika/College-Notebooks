# A quick glance at a few important concepts along with reference material


## RNNs

Recurrent neural networks work like feed forward networks (multi layer perceptrons) but with a twist. Recurrent neural networks unit, takes into account data from past as well. When feed forward networks are concerned with data at current time, RNN cells take into account data from time just before it as well. Usually good with timeseries data.

Saw you have an input data with t vector points, let y be the output. The diagram below shows a simple rnn cell with its unfolding as well

![](https://drive.google.com/uc?export=view&id=1OndrpYKY425EuoESUrp07JHRha2QrCrH)

Thus, RNNs use a kind of internal memory to use information from previous data point into the current data point.

## LSTMs

In the Image above is we are to design a sentence completion work, 
"I live in India. I can speak .....". Suppose you want the prediction to be "Hindi", there is a high chance that simple RNNs may lose that contenxtual information as it can store only short term memory. 

LSTMs, Long Short Term Memory RNNs, are a special kind of RNN which is capable of having long term memories. LSTMs are widely used in Natural Language Processing Space. Thus, LSTMs remember features from last past data. 

Reference material for in-depth understanding
- https://www.analyticsvidhya.com/blog/2021/03/introduction-to-long-short-term-memory-lstm/

## Sequence to Sequence

Seq2Seq modules are used in machine level translation, text summarization, and chatbots. Seq2seq modules involve a encoder decoder network, as you saw in chapter of GANs. Encoder and decoder both are LSTMs, which allow you to translate initial text to some intermediate level and then to a final required level.

## Attention modules

Attention mechanism makes seq2seq modules better and more accurate. As the name suggests, neural network tries to get attentive to a region in the text which has the most relevant information. Basically, it extracts context from the sentence and then does the prediction.

## Transformers

Encoder decoder LSTMs based modules were made better and efficient using transformers. This uses a concept of self-attention mechanism. This self attention system was proposed by a team at Google Brain. When general encoder and decoder systems fail to handle long sequences, the transformers with self attention system help to solve this. When attention maps focus on a certain region of input text, self attention mechanism focusses on the entire text but tries to bring in context into the play. When attention focusses on one part of the data (text), self attention focusses on multiple parts in a single data (text) and tries to find relation between each of those multiple parts. 


Reference material for in-depth understanding

## Bert

BERT - Bidirectional Encoder Representations from Transformers, is one of the most powerful NLP models. The introduction of this module and its variants has changed the way NLP worked, it had made us go one step closer to Artificial General Intelligence. The bidrection enables to take context for a word from the text before it and after it. For example, in the sentence, "They used to go to the school which was next to a train station", so the word "school" can take context from "They used to do to the" as well as from "which was next to a train station". It's a very heavy modules as it has 24 transformer blocks, 16 self attention systems and thus some 340M parameters.

## Using SymSpell

[SymSpellPy](https://github.com/mammothb/symspellpy) is a pythonic wrapper over [SymSpell](https://github.com/wolfgarbe/SymSpell) library, both released as [MIT License](https://github.com/mammothb/symspellpy/blob/master/LICENSE).

This library uses "Symmetric Delete spelling correction algorithm" which is claimed to be **1 million times faster** than traditional algorithms; this algo searches for suggestions from the dictionary way faster as per their paper. 

```python
! pip install -U symspellpy
```

### For single words

```python
# Import module 
import pkg_resources
from symspellpy import SymSpell, Verbosity
```

```python
# Load the language and the model
sym_spell = SymSpell(max_dictionary_edit_distance=2, prefix_length=7)
dictionary_path = pkg_resources.resource_filename(
    "symspellpy", "frequency_dictionary_en_82_765.txt"
)
sym_spell.load_dictionary(dictionary_path, term_index=0, count_index=1)
```

```python
input_term = "stdied" 

suggestions = sym_spell.lookup(input_term, Verbosity.CLOSEST, max_edit_distance=2)

for suggestion in suggestions:
    print(suggestion.term)
```

```python
input_term = "fudied" 

suggestions = sym_spell.lookup(input_term, Verbosity.CLOSEST, max_edit_distance=2)

for suggestion in suggestions:
    print(suggestion.term)
```

# Key terms extraction

The goal of key term extraction is to point out terms in the document that describe the gist (summary) of the document. It could be a word or a phrase or a complete sentence. These key terms are analysed first in the process of natural language understanding. When the documents are huge, these keyterm based analysis becomes very important.

![](https://drive.google.com/uc?export=view&id=1MbPiXsXsSr4LVEw8IGEWvCcqP-jqS8oz)

## Using KeyBERT

[KeyBert](https://github.com/MaartenGr/KeyBERT) is a library based on BERT for handling keyword extraction. It is licensed under [MIT License](https://github.com/MaartenGr/KeyBERT/blob/master/LICENSE). 

![](https://drive.google.com/uc?export=view&id=1ICnj16cQfCMIly-XYeSvEAIzSc0K3qDu)

Uses minimal bert models to find word embeddings. As quoted in the document
- *First, document embeddings are extracted with BERT to get a document-level representation.*
- *Then, word embeddings are extracted for N-gram words/phrases.*
- *Finally, cosine similarity is used to find the words/phrases that are the most similar to the document.* 
- *The most similar words could then be identified as the words that best describe the entire document.*

```python
! pip install keybert
```

```python
# Import library and load model and language
from keybert import KeyBERT

kw_model = KeyBERT()
```

```python
sentence = "You have already seen that data is gathered from multiple places and multiple people. So definitely there is a high chance of bias being introduced in the dataset. For example, say in case of face detection, if the data is being collected in the west there is a high chance that data may have 70% faces of people from west and 30% of faces from east, basically ethic difference. This may create a bias in the AI engine as well as there could be variance in accuracies over face detection based on ethnicity. There are other points of ethical concern here, you need to make sure the data gathered does not violate any copyrights or does not endanger the privacy present in the data. For example, tomorrow you create a dataset for face recognition of your peers and intend to use it only for attendance monitoring system. It is your duty to ensure that the images you gathered are not used for any foul purposes or any other purpose than the intended once. It is your duty to ensure that the data collected has consent from every individual who has agreed to participate."
```

```python
keywords = kw_model.extract_keywords(sentence, keyphrase_ngram_range=(1, 1), 
                                     use_maxsum=True, nr_candidates=10, top_n=10, 
                                     highlight=True)
```

```python
kw_model.extract_keywords(sentence, keyphrase_ngram_range=(1, 4), 
                          top_n=10)
```

# Sentiment Analysis

A sentiment of a text is divided into 3 parts
- positive
- negative
- neutral

This is used by marketing teams in many companies to see whether their marketing texts have positive sentiment or not. Similarly, also used to find sentiment from customer feedback on products and services. 

![](https://drive.google.com/uc?export=view&id=1eY2M_YdU07GuIvtSeaJEpECeflL2QcXH)

## Using SpacyTextBlob

[SpacyTextBlob](https://github.com/SamEdwardes/spacytextblob) is another extension library for spacy, used to extract sentiments from texts. This is licensed as [MIT](https://github.com/SamEdwardes/spacytextblob/blob/main/LICENSE) License. 

```python
! pip install spacytextblob
```

```python
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob
```

```python
nlp = spacy.load('en_core_web_md')
nlp.add_pipe('spacytextblob')
```

```python
sentence = 'The way your customer service takes 1 day to respond back is really frustrating'
doc = nlp(sentence)
print(doc._.blob.polarity)
print(doc._.blob.sentiment_assessments.assessments)
```

```python
sentence = 'Your customer service was quick to respond and they solved my problem promptly'
doc = nlp(sentence)
print(doc._.blob.polarity)
print(doc._.blob.sentiment_assessments.assessments)
```
